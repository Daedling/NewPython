```python
import os
import random
def disgustiingly_simple_human_being(num):
    with open('./sentences/themanhimself/unhomme.txt', 'r', encoding='utf-8') as a:
        stages=a.read().split('$')
    return stages[10-num]
def shred(word):
    res=''
    for a in word:
        res+=a+' '
    return(res[0:len(res)-1])
def cheq(word):
    keepo=''
    while keepo!='N':
        notr=11
        letterslist=list('ёйцукенгшщзхъфывапролджэячсмитьбю')#capitals reserved for shenanigans                                             
        word=list(word.lower())                               #also counts all attempted letters
        out=('_'*len(word)).capitalize()
        print(shred(out)+'\n'+'У вас '+ru(notr)[1])
        while notr!=0:#attempts check
            iput=input('letter: ').lower()
            if iput in word:  #checks for letter coincidence      
                if iput in letterslist:#checks if first letter instance
                    out=''              #output template
                    letterslist.remove(iput)   #for checks if is in letter pool
                    for a in word:  #filling output based on missing letters
                        if a not in letterslist:
                            out+=a         
                        else:
                            out+='_'
                    if list(out)==word:     #winning condition
                        print('congratulations\n'+out.capitalize())

                        
                        break
                else:                           #checks if letter is used already
                    print('Буква уже использована')
                    continue
            else:
                if iput not in letterslist:#shenanigans to avoid making another list
                    if iput.capitalize() not in letterslist:
                        print('Нет такой буквы!')
                        continue
                    else:
                        print('Символ уже использован')
                        continue
                else:
                        notr-=1
                        letterslist.remove(iput)#more shenanigans
                        letterslist.append(iput.capitalize())
                        print(disgustiingly_simple_human_being(notr))
                        print('Неверная буква')
            print(shred(out.capitalize())+'\n'+ru(notr)[0])
        if notr==0:
            print('he is dead, nice try nonetheless')
        while keepo !='Y' or keepo != 'N':
            keepo=input('Хотите начать заново? Y/N: ')
            if keepo == 'Y':
                cheq(wordchoice())
                return
            if keepo == 'N':
                print('Хорошего дня')
                return

def wordchoice():
    print('Список тем: ')
    a=os.listdir('./sentences')
    for  ct,b in enumerate (a):
        if ct == 0:
            continue
        a[a.index(b)]=b[:len(b)-4]
        print(' * '+a[ct])
    c=input('Введите название темы из списка: ').capitalize()
    while c not in a:
        c=input('Такой темы в списке нет, попробуйте еще раз: ').capitalize()
    print('Вы выбрали тему: '+c)
    with open('./sentences/'+c+'.txt', 'r',encoding='utf-8') as k:
        return(random.choice(k.read().split('\n')[0].split('*')))
def ru(notr):
    a=str(notr)[len(str(notr))-1]
    if int(notr)<=20 and int(notr) >=11:
        return 'Осталось '+str(notr)+' попыток',str(notr)+ ' попыток'
    else:
        if a=='1':
            return 'Осталась '+str(notr) +' попытка',str(notr)+ ' попыток'
        if a =='2'or a == '3' or a =='4':
            return 'Осталось '+str(notr) +' попытки',str(notr)+ ' попыток'
        else:
            return 'Осталось '+str(notr)+ ' попыток',str(notr)+ ' попыток'

cheq(wordchoice())
```
