```python
from flask import Flask,redirect,render_template,request
import re
app = Flask(__name__)
@app.route('/processor')
def proc():
    with open ('result.csv','a',encoding='utf-8')as k:
        if request.args['firstname']=='':
            k.write('\n'+'None;')
        else:
            k.write('\n'+request.args['firstname']+';')
        if request.args['lastname']=='':
            k.write('None;')
        else:
            k.write(request.args['lastname']+';')
        k.write(request.args['age']+';')
        k.write(request.args['word1']+';')
        k.write(request.args['word2']+';')
        k.write(request.args['word3']+';')
        k.write(request.args['word4']+';')
        k.write(request.args['word5']+';')
        k.write(request.args['word6']+';')
        k.write(request.args['word7'])
    return  '''<br><a href="/form" class="button">Form link</a>
   <br><a href="/search" class="button">Search again</a>
    <br><a href="/stats" class="button">Results link</a>'''
@app.route('/stats')
def results():
    cor=0
    wro=0
    outnum=3
    outcome=''
    with open ('result.csv', 'r', encoding='utf-8') as k:
        k=k.read().split('\n')
        while outnum<=9:
            for num, a in enumerate(k):
                a=a.split(';')
                if num!=0:
                    print(a,outnum)
                    if int(a[outnum])-10<10:
                        cor+=1
                    else:
                        wro+=1
            outcome+='results for: '+k[0].split(';')[outnum]+' - correct:'+str(cor)+' wrong:'+str(wro)+'<br>'
            outnum+=1
            wro=0
            cor=0
        for num, a in enumerate(k):
            a=a.split(';')
            if num!=0:
                outcome+=a[0]+' '+a[1]+'; age '+a[2]+'<br>'
    outcome+='<br><a href="/form" class="button">Form link</a>'
    outcome+='<br><a href="/search" class="button">Search again</a>'
    outcome+='<br><a href="/stats" class="button">Results link</a>'
    return outcome    
            
@app.route('/')
def form():
    return render_template('test.html')
@app.route('/search')
def wrg():
    return(render_template('search.html'))
           
@app.route('/results')
def index():
    outcome=''
    with open('result.csv', 'r', encoding='utf-8') as k:
        k=k.read().split('\n')
        for num, a in enumerate(k):
            a=a.split(';')
            if num!=0:
                outcome+=a[0]+', '+a[1]+': '
                if int(a[1+int(request.args['question'])])-10<10:
                    outcome+='Correct<br>'
                else:
                    outcome+='Wrong<br>'
    outcome+='<br><a href="/form" class="button">Form link</a>'
    outcome+='<br><a href="/search" class="button">Search again</a>'
    outcome+='<br><a href="/stats" class="button">Results link</a>'
    return outcome



if __name__ == '__main__':
    app.run(debug=False)
```
